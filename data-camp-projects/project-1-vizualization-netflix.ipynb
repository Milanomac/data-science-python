```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("datasets/the_office_series.csv")
data
```

### Problem 1

```python
# Scaling the ratings
data['scaled_ratings'] = data['Ratings'] / 10
data.head()
```

```python
# Color system using list
colors = []
for index, row in data.iterrows():
    if row['scaled_ratings'] < 0.7:
        colors.append('red')
    elif row['scaled_ratings'] < 0.8:
        colors.append('orange')
    elif row['scaled_ratings'] < 0.9:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
print(colors[0:10]) #printing out a few results to see if the loop works well
```

```python
#  Sizing system
# Creating a column with boolean values to see if the episode has guests or not
import numpy as np
data['has_guests'] = np.where(data['GuestStars'].str.contains('NaN'), 0, 1)
data
```

```python
sizes = []

for index, row in data.iterrows():
    if row['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)
sizes[0:10] #printing out a few results to see if the loop works well
```

```python
# Plotting 

# Create 'episode_number' variable with actual episode numbers
data['episode_number'] = data['Unnamed: 0'] + 1 
```

```python
plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('ggplot')
fig, ax = plt.subplots()

ax.scatter(x=data['episode_number'],
           y=data['Viewership'], 
           c=colors, 
           s=sizes,
           marker='*')

ax.set_xlabel("Episode Number")
ax.set_ylabel("Viewership (Millions)")
ax.set_title("Popularity, Quality, and Guest Appearances on the Office")

plt.show()
```
